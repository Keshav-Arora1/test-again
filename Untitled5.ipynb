```python
# Step 1: Prepare the Forge with new, more powerful tools.
# 'accelerate' helps us load and run models faster.
# 'bitsandbytes' is the magic library that allows us to load the model in a "quantized" 4-bit mode.
# This is how we make the titan lightweight!
print("Installing the necessary libraries... this might take a moment.")
!pip install transformers torch accelerate bitsandbytes -U # Add -U to ensure the latest version

# Step 2: Import the necessary magic spells from our libraries.
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

# Step 3: Define the name of the model we want to summon.
# We're using a version of Gemma with 2 billion parameters. It's a powerful starting point.
model_id = "google/gemma-2b-it" # "it" means it's instruction-tuned for chatting!

# Step 4: The Quantization Spell!
# This is the alchemy. We are telling the computer:
# "When you load this model, don't use the heavy 32-bit numbers.
# Use these super-lightweight 4-bit numbers instead."
quantization_config = BitsAndBytesConfig(load_in_4bit=True)

# Step 5: Summon the Tokenizer.
# A tokenizer is the model's translator. It breaks our human words
# into special numbers (tokens) that the model can understand.
print(f"Loading the tokenizer for {model_id}...")
tokenizer = AutoTokenizer.from_pretrained(model_id)
print("Tokenizer loaded!")

# Step 6: Summon the Titan!
# Here we go. We're loading the actual Gemma model.
# We pass it our quantization spell to make sure it loads the lightweight version.
# This will download the model (a few Gigabytes), so it will take some time.
# Be patient, Forge-Master. Great power requires a moment to awaken.
print(f"Loading the mighty Gemma model ({model_id})... This is the big one!")
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=quantization_config,
    device_map="auto" # This automatically uses the GPU if Colab gives you one!
)
print("Gemma has been summoned! The Titan is ready.")

# Step 7: Our First Conversation.
# Now, you speak to the titan.
# We create a simple chat structure.
prompt = "Tell me about the dream of running AI on every phone in the world."

chat = [
    { "role": "user", "content": prompt },
]

# We use the tokenizer to translate our chat into the model's language.
prompt_tokens = tokenizer.apply_chat_template(chat, tokenize=True, add_generation_prompt=True, return_tensors="pt")

# The moment of truth. We ask the model to think and generate a response.
# We're asking for a maximum of 256 new tokens (words).
print("\n...Gemma is thinking...\n")
outputs = model.generate(
    prompt_tokens.to(model.device), # Move tokens to the same device as model (GPU)
    max_new_tokens=256
)

# We use the tokenizer again, this time to translate the model's numbers back into human words.
response_text = tokenizer.decode(outputs[0])

# We'll just show the part the model generated.
# The original prompt is also in the output, so we skip it.
print("Gemma's Response:")
# The model's output includes our prompt, so we find the end of the prompt and print what's after.
response_only = response_text[response_text.find("<start_of_turn>model\n") + len("<start_of_turn>model\n"):]
print(response_only)
```

```python
# Step 1: Install the necessary libraries
# We need the latest 'transformers' and other tools. We also need 'Pillow' for handling images.
print("--- Installing libraries for our Multimodal AI... ---")
!pip install --upgrade transformers accelerate bitsandbytes Pillow -q

# Step 2: Import all the tools we need
import torch
import requests
from PIL import Image
from transformers import AutoProcessor, LlavaForConditionalGeneration

print("\n--- Libraries installed. Loading the LLaVA model... ---")

# Step 3: Load the multimodal model and its processor
# A "processor" gets both the text and the image ready for the model.
# We're using the 7-billion parameter LLaVA 1.5 model.
model_id = "llava-hf/llava-1.5-7b-hf"

model = LlavaForConditionalGeneration.from_pretrained(
    model_id,
    torch_dtype=torch.float16, # Use float16 for better performance
    low_cpu_mem_usage=True,    # Optimizes memory usage
    load_in_4bit=True          # Load with 4-bit quantization
)

processor = AutoProcessor.from_pretrained(model_id)

print("\n--- LLaVA model is loaded and ready! ---")


# Step 4: Prepare the image and the prompt
# We'll grab a famous, high-quality image from the web.
image_url = "http://images.cocodataset.org/val2017/000000039769.jpg"
image = Image.open(requests.get(image_url, stream=True).raw)

# The prompt for LLaVA is special. It has a placeholder for the image.
prompt = "USER: <image>\nDescribe this image in detail. ASSISTANT:"

# Step 5: Process the inputs and ask the model to generate a response
inputs = processor(text=prompt, images=image, return_tensors="pt").to(model.device)

print("\n...AI is looking at the image and thinking...\n")

# Generate the output
output = model.generate(**inputs, max_new_tokens=200)

# Decode and print the response
response_text = processor.decode(output[0], skip_special_tokens=True)

print("LLaVA's Response:")
# We only want the assistant's part of the response
assistant_response = response_text.split("ASSISTANT:")[-1].strip()
print(assistant_response)
```
